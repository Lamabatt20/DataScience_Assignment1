<span style="color:red"> **Create a new column called unified_price that represents the price in USD per kilogram for commodities measured by weight and the price per liter for commodities measured by volume.** </span>

```python
# Custom function to calculate  unified_price
def unified_price(row):
    #split unit column for each row 
    Value_unit= row['unit'].split()
    value = 1.0  
    if len(Value_unit) > 1 and Value_unit[0].isdigit():
        value = float(Value_unit[0]) 
    elif len(Value_unit) == 1 and Value_unit[0].isdigit():
        value = float(Value_unit[0])
        
    if 'KG' in row['unit']:
        return row['usdprice'] /value
    elif 'G' in row['unit']:
        return row['usdprice']  /(value/ 1000 )
    elif 'L' in row['unit']:
        return row['usdprice'] /value 
    elif 'Cubic meter' in row['unit']:
        return row['usdprice'] /(value * 1000 )
    else:
        return None
# Applying the function to each row
df['unified_price'] = df.apply(unified_price, axis=1)
df
```

<span style="color:blue">This is the dataset after we calculated the unified price, so we created a new column called unified_price and the results were as follows: price in USD per kilogram, price per liter for each commodity measured by weight or volume
  </span>

<span style="color:red"> **Using the unified_price, find the maximum price reached during this period and for which commodity for both:
commodities measured by weight
commodities measured by volume** </span>

```python
#find the maximum price for each commodity measured by weight
max_unified_price = df[(df['unit'] == 'KG') | (df['unit'] == 'G')]['unified_price'].max() 
maximum_weight = df[df['unified_price'] == max_unified_price].iloc[0]
print(f"Commodity: {maximum_weight['commodity']}, Unified Price: {maximum_weight['unified_price']}")
```

<span style="color:blue">This is the result of the maximum price commodity measured by weight, which is Meat (goat, with bones),and its value Unified Price: 17.3144 which is the maximum price of commodities measured by weight.
  </span>

```python
#find the maximum price for each commodity measured by volume
max_unified_price = df[(df['unit'] == 'L') | (df['unit'] == 'Cubic meter')]['unified_price'].max()
maximum_volume = df[df['unified_price'] == max_unified_price].iloc[0]
print(f"Commodity: {maximum_volume['commodity']}, Unified Price: {maximum_volume['unified_price']}")
```
